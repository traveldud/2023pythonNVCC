#### file operation without needing to close (it's automated)

```python
#with combined with 'open' it's interactive as 'fo' file operation (fo - can be any acronym)
#normally you always need to close a file that's been opened, python  does it 
#automatically when you finish your executable when you use the 'with ' operator

with open("gutenberg_shakespeare.txt", "r") as fo:  #"r", "w", and others...
    for line in fo:
        print(line.strip())
```

#### method were closing of the file is needed -- not using 'with'

```python
#Without using with operator
fo = open("gutenberg_shakespeare.txt", "r")

for line in fo:
    print(line.strip())
fo.close() #perform close exclusively
```

### Use with keyword to write  ta file

```python
with open("output.txt","w") as file:
    file.write("Hello World")
```

### Append to an existing file


```python
with open("output.txt","a") as file:
    file.write("Appended Text")
```

### Copying a file

```python
with open("gutenberg_shakespeare.txt", "r") as source:
    content = source.read()
    with open("destination.txt", "w") as destination:
        destination.write(content)
```
