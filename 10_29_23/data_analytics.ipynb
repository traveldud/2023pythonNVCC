### Anonymous function
Use lambda keyword to write your anonymous function

```python
def add(x,y):
    return x + y
```

```python
add(2,3)
```

```python
addlambda = lambda x,y: x + y + add(x,y) #anonymous INLINE function within in the module 
            #( reason! -> 'local' or 'private')
            #you can call other functions from this inline functions
addlambda(2,3)
```

```python
add3 = lambda x,y,z: x+y-z
add3(1,2,3)
```

```python
#square a Number
square = lambda x: x**2
square(4)
```

```python
#Concatenate Two Strings
concatenate = lambda a,b: a + b
concatenate("hello ", "world!!")
```

```python
# Check if a number is even
is_even = lambda x: x%2 == 0
is_even(3)
```

```python
is_even(4)
```

```python
#Determine the Maximum of three numbers
maximum = lambda x, y, z:  x if x > y and x > z else (y if y > z else z)
maximum(1,5,3)
```

```python
# Calling a lambda Function inside a for loop
numbers = [1,2,3,4]
square = lambda x: x**2
for num in numbers:
    print(square(num))
```

```python
numbers = [1,2,3,4,5]
for num in numbers:
    print((lambda x: x**2)(num))
    print(num**2)
```

### Analyzing Chipotle Data


```python
import pandas as pd
import collections
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
url = "https://raw.githubusercontent.com/braincomputingsantosh/Data/master/chipotle.tsv.txt"
chipoDF = pd.read_csv(url, sep= '\t') # this is a '\t' - tab separated data, 
                        #when only ',' the "sep = ',' is not listed, ie. *(url)"
chipoDF
```

```python
chipoDF.head(10)
```

```python
chipoDF.shape[0]
```

```python
chipoDF.info()
```

```python
c = chipoDF.groupby('item_name')
c = c.sum()
c = c.sort_values(['quantity'], ascending=False)
```

```python
c.head()
```

```python
#create an anonymous lambda function called dollarizer
dollarizer = lambda x: float(x[1:-1])
chipoDF.item_price = chipoDF.item_price.apply(dollarizer) #sending it to the dollarizer function
chipoDF.head()
```

```python
dollarizer = lambda x: str(x[2:-1])
chipoDF.item_name = chipoDF.item_name.apply(dollarizer) #sending it to the dollarizer function
chipoDF.head()
```

```python
### Create a Historgram of the top 5 items bought
x = chipoDF.item_name

letter_counts = collections.Counter(x)

df = pd.DataFrame.from_dict(letter_counts, orient='index')
df = df[0].sort_values(ascending = True)[45:50]
#df - df[0].sort_values(ascending = True)[45:50]

df.plot(kind='bar')

plt.xlabel('Items')
plt.ylabel('Quantity')
plt.title('Most Ordered Items')

```
